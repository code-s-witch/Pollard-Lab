```python

wdir= '../../../data/projects/morin/'


#create a dictionary that follows the order of enhancers.bed
inFile = open(wdir+'proseq_intergenic_counts.csv', 'r')
l2f_change = {}
for line in inFile:
    splitLine = line.strip().split('\t')
#     print(splitLine)
    chrom = splitLine[0]
    start =  splitLine[1]
    end = splitLine[2]
    key = chrom + ':' + start + ':'+ end
    start_list = []
    l2f_change[key] = start_list     
inFile.close()


inFile = open(wdir+'proseq_overlapping_enhancers_filtered.bed', 'r')
for line in inFile:
    splitLine = line.strip().split('\t')
#     print(splitLine)
    chrom = splitLine[0]
    start =  splitLine[1]
    end = splitLine[2]
    key = chrom + ':' +  start + ':'+ end
    val = float(splitLine[3])
    if l2f_change.get(key)!=None:
        #check if key exists, if it does append to it
        newList = l2f_change[key]
        newList.append(val)
        l2f_change[key] = newList
inFile.close()
#fill empty lists with 0        
for key, val in l2f_change.items():
    if not val:
        l2f_change[key] = [0]        


# print(l2f_change.items())

#for enhancer and list of l2f change values, avg the values
for e,l in l2f_change.items():
#     print(len(l))
    sum_ = 0
    for x in range(0,len(l)):
        sum_ += l[x]     
    avg = sum_/len(l)
    l2f_change[e] = avg
    
# print(l2f_change.items())


# print(len(l2f_change.keys()))
#since the dictionary is unordered we have to order it so we can create a 1:1 correspondence with enhancer.bed
outFile = open(wdir+'enhancers_log2foldchange.csv', 'a+')
for key, l2f in l2f_change.items():
    splitLine = key.split(':')
    chrom = splitLine[0]
    start = splitLine[1]
    end = splitLine[2]
#     print(key.split(':'))
    line = chrom +'\t'+ start + '\t' + end + '\t' + str(l2f)+ '\n'
#     print(line)
    outFile.write(line)
outFile.close()
# print(len(l2f_change.keys()))


```
